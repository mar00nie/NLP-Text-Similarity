```python
# Load libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk, re, string
import pickle

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
def read_text_file(file):
    
    '''Reads lines in file and appends to a corpus list'''
    
    corpus = []
    for i, line in enumerate(file):
        if i==0:
            continue
        else:
            corpus.append(str(line))
    
    return corpus
```

```python
billwords = open("1-billion-word-language-modeling-benchmark-r13output.tar/1-billion-word-language-modeling-benchmark-r13output.tar", "r", encoding="ISO-8859-1")
```

```python
billwords_corpus = read_text_file(billwords)
```

```python
billwords_corpus
```

```python
billwords_df_small = pd.DataFrame(billwords_corpus[0:100000], columns=['text'])
```

```python
billwords_df_small
```

```python
def preprocess_text(row):
    
    '''Cleans up corpus data used to train NLP models'''
    
    translator = str.maketrans("", "", string.punctuation)
    
    edit = row['text'].translate(translator)
    edit = edit.lower().strip()
    #re.sub(' +', ' ', line)
    edit = [edit.split()]
    
    return edit
```

```python
billwords_df_small['edited'] = billwords_df_small.apply(preprocess_text, axis=1)
```

```python
billwords_df_small
```

```python
billwords_prep_small = billwords_df_small['edited'].tolist()
```
